```python
import pandas as pd

PATH = 'source/'
matricula = pd.read_csv(PATH +'matriculas-2022.2.csv',sep=";",decimal=",")
turmas = pd.read_csv(PATH +'turmas-2022.2.csv',sep=";",decimal=",")
componentes_curriculares = pd.read_csv(PATH +'componentes-curriculares-presenciais.csv',sep=";",decimal=",")
cursos = pd.read_csv(PATH +'cursos-ufrn.csv',sep=";",decimal=",")
```

```python
import inspect
def retrieve_name(var):
    callers_local_vars = inspect.currentframe().f_back.f_locals.items()
    return [var_name for var_name, var_val in callers_local_vars if var_val is var]
```

```python
def criterio_nota_media(df,nota=3,media=5):
    situacao = None
    menor_que_min = False
    if df['nota'].min() < nota:
        menor_que_min = True
    if df['nota'].mean() < nota:
        situacao = 'Reprovação'
    elif menor_que_min and df['nota'].mean() >= 7:
        situacao = 'Aprovação'
    elif df['nota'].mean() >=  media:
        situacao = 'Aprovação'
    else:
        situacao = 'Reposição'
    return pd.Series([situacao, menor_que_min], index=['Situação', 'Tirou nota mínima ?'])
```

```python
descricao_usadas = ['APROVADO','REPROVADO POR MÉDIA E POR FALTAS','APROVADO POR NOTA', 'REPROVADO','REPROVADO POR FALTAS', 'REPROVADO POR NOTA E FALTA','REPROVADO POR NOTA']
analise_reprovacao = matricula_turma.loc[matricula_turma['descricao'].isin(descricao_usadas)]
analise_reprovacao = analise_reprovacao.loc[analise_reprovacao['unidade'] <= 3,['id_turma','discente','unidade','nota']]
situacao_atual = analise_reprovacao.groupby(['id_turma','discente']).apply(criterio_nota_media)
proposta = analise_reprovacao.groupby(['id_turma','discente']).apply(lambda x: criterio_nota_media(x,nota=4,media=6))
```

```python
medias = analise_reprovacao[['nota','unidade']]
def thousands_formatter(x, pos):
    return f'{round(x / 1e3)} mil'

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Frequência das notas", fontsize=18)
ax.set_xlabel("Nota", fontsize=12)
ax.set_ylabel("Frequência", fontsize=12)
ax.hist(medias['nota'],bins=10,ec='white')
formatter = FuncFormatter(thousands_formatter)
ax.yaxis.set_major_formatter(formatter)
ax.xaxis.set_ticks(np.arange(11))
plt.show()
```
